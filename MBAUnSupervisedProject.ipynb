```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from apyori import apriori
```

```python
CustomerData = pd.read_csv('/content/drive/MyDrive/UnSupervisedMachineLearningData/customers.csv',sep=',')
```

```python
ArticlesData = pd.read_csv('/content/drive/MyDrive/UnSupervisedMachineLearningData/articles.csv',sep=',', dtype={'article_id': str , 'prod_name' : str})
```

```python
ArticlesData = ArticlesData[['article_id', 'prod_name','product_type_name','perceived_colour_value_name','section_name','garment_group_name']]
```

```python
ArticlesData['Combined'] = ArticlesData[['prod_name', 'product_type_name', 'perceived_colour_value_name','section_name','garment_group_name']].apply(lambda x: ', '.join(map(str, x)), axis=1)
```

```python
articles_dict = dict(zip(ArticlesData['article_id'], ArticlesData['Combined']))
```

```python
TransactionDataFrame = TransactionData.drop(columns=["price","sales_channel_id"])
```

```python
totalbaskets = Aprior_Transacitons['article_id'].tolist()
```

```python
association_rules = list(apriori(totalbaskets, min_support=0.0005, min_lift=200))
```

```python
for rule in association_rules:
    print("Items: " + str(rule.items))
    print("Support: " + str(rule.support))
    print("Confidence: " + str(rule.ordered_statistics[0].confidence))
    print("Lift: " + str(rule.ordered_statistics[0].lift))
    print("\n")
```

```python
for rule in association_rules:
    listofitems = list(rule.items)
    print("Items: " + str(list(map(map_item_to_dict, listofitems))) )
    print("Support: " + str(rule.support))
    print("Confidence: " + str(rule.ordered_statistics[0].confidence))
    print("Lift: " + str(rule.ordered_statistics[0].lift))
    print("\n")
```

# FP Growth Algorithm

```python
te = TransactionEncoder()
te_ary = te.fit(totalbaskets).transform(totalbaskets)
transformedDataFrame = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
transformedDataFrame
```

```python
frequent_itemsets = fpgrowth(transformedDataFrame, min_support=0.00050, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2.0)
```

```python
rules = pd.DataFrame(rules)
```

```python
def map_items_to_ids(items):
    return [articles_dict[item] for item in items]

rules['antecedents'] = rules['antecedents'].apply(map_items_to_ids)
rules['consequents'] = rules['consequents'].apply(map_items_to_ids)
```

```python
rules.sort_values(by='lift', ascending=False, inplace=True)
```
